# Numpy

```python
import numpy as np
```

NumPy supports vectorized operations such as element-wise addition, subtraction etc which is not the case with basic lists.

```python
lst = [5, 10, 0, 200] 
arr = np.array(lst)
print(arr + 5)
arr
```

A Heterogeneous python list is comparitvely slower than numPy nd-array because python has to store type information for each element however numpy works with similar datatype and is faster in processing time

```python
lst = [1, 2, 3, 'text', True, 3+2j] 
import numpy as np
arr = np.array(lst)
print(type(arr[0]),type(arr[4]), type(arr[5]))
```

Python list accounts for high memory size due to storing information for each pointer as well as to an object they are pointing to. However, NumPy array stores uniform values.

```python
lst = [56, 45, 12, 6]
import numpy as np
arr = np.array(lst)
print("size of numpy array: ",arr.nbytes)

import sys
print("size of equivalent python list",sys.getsizeof(lst))
```

## Vectorized Operations

```python
X = np.array([[5, 6, 7, 8], [9, 8, 7, 6]])
print(X)

Y = np.array([[1, 2, 3, 4,], [2, 3, 4, 5]])
print(Y)
```

```python
print(X-Y)
```

```python
print(X < 7)
```

```python
# another way to run above code [print(X < 7)]
sevens_mat = np.ones((2, 4)) * 7
print(np.less(X, sevens_mat))
```

## Image

```python
import os.path
from skimage.io import imread
rom skimage import data_dir
img = imread(os.path.join(data_dir, 'chelsea.png')) 

# Display the image
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
# Slicing the eyes
img_slice = img.copy()
img_slice = img_slice[70:170,120:350]
plt.imshow(img_slice)
```

```python
# Horizontal flip
plt.imshow(np.fliplr(img_slice))
```

```python
# Vertical flip
plt.imshow(np.flipud(img_slice))
```

```python
# Rotate by 90 degree
plt.imshow(np.rot90(img_slice))
```
